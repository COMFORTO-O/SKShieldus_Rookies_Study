```python
import requests
import bs4
from bs4 import BeautifulSoup
from urllib.parse import urljoin


section_dict = {
    'recent_news' : 'recent?mid=n0100',
    'politics' : 'section?mid=n0200',
    'econmy' : 'section?mid=n0300',
    'society' : 'section?mid=n0400',
    'world' : 'section?mid=n0500',
    'it_science' : 'section?mid=n0600',
}

def print_news(section_name):
    section = section_dict.get(section_name)
    url = f'https://news.nate.com/{section}'
    print(f"{url} {section_name} 뉴스")

    res = requests.get(url)
    res.encoding = 'euc-kr'
    soup = BeautifulSoup(res.text, 'html.parser')


    news_items_list = soup.select_one('div#newsContents')
    news_items = news_items_list.select('div.mduCluster')

    for div_tag in news_items:
        a_tag = div_tag.find('a')
        if a_tag is None:
            continue
        link = a_tag.get('href', '')
        title = div_tag.select_one('h2.tit')

        img_tag = div_tag.find('img')
        if img_tag and img_tag.get('src'):
            src = img_tag['src']
            full_url = urljoin('https:', src)
            print("이미지 URL:", full_url)
        else:
            print("이미지를 찾을 수 없습니다.")

        print(link)
        print(title.get_text(strip=True))
    print()

print_news('politics')
print_news('world')
```

```python
import os
import requests
from bs4 import BeautifulSoup
from urllib.parse import urljoin

def download_one_episode(title, no, url):
    # 저장 경로 설정
    folder_path = os.path.join("img", title, str(no))
    os.makedirs(folder_path, exist_ok=True)
    print(f"저장 경로: {folder_path}")

    headers = {
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/135.0.0.0 Safari/537.36"
    }

    # 웹툰 페이지 요청
    response = requests.get(url, headers=headers)
    soup = BeautifulSoup(response.text, "html.parser")

    # 웹툰 이미지 추출
    image_container = soup.select_one('div.view_area > div.wt_viewer')
    if not image_container:
        print("이미지 영역을 찾을 수 없습니다.")
        return

    images = image_container.find_all('img')
    print(f"이미지 {len(images)}개 발견됨.")

    for idx, img_tag in enumerate(images):
        img_url = img_tag.get("src")
        if not img_url:
            continue

        # 절대 URL 처리
        img_url = urljoin(url, img_url)

        # 이미지 다운로드
        try:
            img_data = requests.get(img_url, headers=headers).content
            img_filename = os.path.join(folder_path, f"{idx + 1:03}.jpg")
            with open(img_filename, "wb") as f:
                f.write(img_data)
            print(f"저장됨: {img_filename}")
        except Exception as e:
            print(f"이미지 다운로드 실패: {e}")

# 예시 호출
download_one_episode('낢이사는이야기', 48, 'https://comic.naver.com/webtoon/detail?titleId=833255&no=50&week=tue')
```
